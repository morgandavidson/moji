### UN geographic categories as an [edge list](https://en.wikipedia.org/wiki/Edge_list)

```python
import pandas as pd
df = pd.read_excel('../../Data_sources/World/WPP2022_GEN_F01_DEMOGRAPHIC_INDICATORS_COMPACT_REV1.xlsx',header=16)
df = df.iloc[:,[2,4,9]]
df = df.drop_duplicates()
df = df.reset_index(drop=True)
df.columns = ['child','child_code','parent_code']
df['parent'] = df['parent_code'].map(df.set_index('child_code')['child'])
df.head(3)
```

### UN geographic categories as a tree

```python
from anytree import Node, RenderTree
def add_nodes(nodes, parent, child):
    if parent not in nodes:
        nodes[parent] = Node(parent)  
    if child not in nodes:
        nodes[child] = Node(child)
    nodes[child].parent = nodes[parent]

nodes = {}  # store references to created nodes 
# df.apply(lambda x: add_nodes(nodes, x["Parent"], x["Child"]), axis=1)  # 1-liner
for parent, child in zip(df["parent"],df["child"]):
    add_nodes(nodes, parent, child)

roots = list(df[~df["parent"].isin(df["child"])]["parent"].unique())
for root in roots:         # you can skip this for roots[0], if there is no forest and just 1 tree
    for pre, _, node in RenderTree(nodes[root]):
        print("%s%s" % (pre, node.name))
```
